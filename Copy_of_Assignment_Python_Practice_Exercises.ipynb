**Guessing Game**

If the user's unable to guess in 3 attempts, show **"Sorry you failed!"**
If the user guess it right, show **"You win!"** and game ends.

````python
secret_number=9
attempts_left=3
while attempts_left>0:
 guess=int(input("enter the number"))
 if guess==secret_number:
  print("You win")
  break
 else:
  attempts_left-=1
  print(f"Wrong guess you have {attempts_left} attempts left")
if attempts_left==0:
 print("Sorry you failed")
````
